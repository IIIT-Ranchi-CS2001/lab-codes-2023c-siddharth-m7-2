Check if the file `election_data.csv` exists in the directory. If not, create the file and write the election data into it. Handle file-related exceptions gracefully.

```python
import os
import csv
if not os.path.exists('election_data.csv'):
    with open('election_data.csv', 'w', newline='') as csvfile:
        csvwriter = csv.writer(csvfile)
        csvwriter.writerow(['State', 'Party', 'SeatsWon','Total_Seats','Voter_Turnout(%)'])
        csvwriter.writerow(['MadhyaPradesh', 'BJP', 163, 230, 72.1])
        csvwriter.writerow(['MadhyaPradesh', 'INC', 66, 230, 72.1])
        csvwriter.writerow(['MadhyaPradesh', 'BSP', 0, 230, 72.1])
        csvwriter.writerow(['MadhyaPradesh', 'Others', 1, 230, 72.1])
        csvwriter.writerow(['Rajasthan', 'BJP', 115, 200, 74.2])
        csvwriter.writerow(['Rajasthan', 'INC', 69, 200, 74.2])
        csvwriter.writerow(['Rajasthan', 'BSP', 2, 200, 74.2])
        csvwriter.writerow(['Rajasthan', 'Others', 13, 200, 74.2])
        print('File `election_data.csv` created.')
else:
    print('File `election_data.csv` already exists.')


```

Read the data into a Pandas DataFrame and calculate the percentage of seats won by each party. Add this as a new column named Seats_Percentage.


```python
import pandas as pd
df = pd.read_csv('election_data.csv')
df['Seats_Percentage'] = df['SeatsWon'] / df['Total_Seats'] * 100
print(df)

```

Determine the party with the highest number of seats in each state and display their names.


```python
df1 = df.groupby('State')['SeatsWon'].idxmax()
print(df.loc[df1, ['State', 'Party', 'SeatsWon']])

```

Create a bar chart showing the number of seats won by each party in each state using Matplotlib


```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('election_data.csv')
df1 = df.pivot(index='State', columns='Party', values='SeatsWon')
df1.plot(kind='bar')
plt.ylabel('Seats Won')
plt.title('Seats Won by Each Party in Each State')
plt.show()
```
